```python
import matplotlib.pyplot as plt
import random
```

```python
benchmark = ["401.bzip","409.mcf","456.hmmer","458.sjeng","470.libm"]
Execution_Time = [83.654,62.553,59.390,513.823,174.763]
CPI =[1.673085,1.25106,71.187803,10.276466,3.495270]
L1_dCache_MissRate = [1.4312 , 0.2062 , 0.1628 , 12.831 ,6.0972]
L1_iCache_MissRate = [0.0075 , 1.9032 , 0.0212 , 0.0020 , 0.0095 ]
L2_Cache_MissRate =  [29.5247 , 6.7668 , 7.8296 , 99.9978 , 99.9940 ]

stats_array = [Execution_Time , CPI , L1_dCache_MissRate , L1_iCache_MissRate , L2_Cache_MissRate]

labels = ["Execution Time (ms)","CPI" , "L1 dCache MissRate (%)" , "L1 iCache MissRate (%)", "L2 Cache MissRate (%)"]
```

```python
import random
for i in range(len(stats_array)):
    plt.figure(figsize=(16,10))
    plt.figure(i+1)
    plt.ylabel(labels[i])
    plt.bar(benchmark,stats_array[i],color=(random.random(),random.random(),random.random(),1),edgecolor=("black"))
    plt.savefig("C:/Users/Alexandros/Desktop/gem5images/{}".format(labels[i])) 
plt.show()    
```
